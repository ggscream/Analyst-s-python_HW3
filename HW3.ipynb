```python
# Задача 1.1
import pandas as pd
df = pd.read_csv('./kc_house_data.csv')
df.head()
```

```python
# Задача 1.2
def dt_renovated(y,x):
    if y == 0: return 0 
    else: return y - x
df['delta_renovated'] = df.apply(lambda x: dt_renovated(x['yr_renovated'],x['yr_built']), axis=1)
df.head()
```

```python
# Задача 1.3
def year_sale(x):
    return str(x)[:4]
def month_sale(x):
    return str(x)[4:6]
df['year_of_sale'] = df['date'].apply(year_sale)
df['month_of_sale'] =  df['date'].apply(month_sale)
df.head()
```
